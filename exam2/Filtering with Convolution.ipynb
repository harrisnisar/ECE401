```python
%matplotlib inline 

import numpy as np
from PIL import Image
from matplotlib import pyplot as plt
import h5py
```

```python
plt.style.use('seaborn')
```

```python
with h5py.File('data.hdf5','r') as f:
    print(list(f.keys()))
    onesecond = f['onesecond'][:]
    sfreq = f['sfreq'][0]

print('According to data.hdf5, sfreq is %g, nchannels is %d, nsamps is %d'%
     (sfreq, onesecond.shape[0], onesecond.shape[1]))
```

Data is collected from 60 EEG channels. We have 1 second of that data which is 600 samples long (Fs=600/1).

```python
onesecond.shape
```

```python
onesecond_channel_0 = onesecond[0,:]
onesecond_channel_30 = onesecond[30,:]
onesecond_channel_59 = onesecond[59,:]
```

```python
fig, axs = plt.subplots(3)
axs[0].plot(onesecond_channel_0)
axs[0].set_title("1 sec of channel 0")

axs[1].plot(onesecond_channel_30)
axs[1].set_title("1 sec of channel 30")

axs[2].plot(onesecond_channel_59)
axs[2].set_title("1 sec of channel 59")

fig.tight_layout()

plt.show()
```

```python
smoothing_filter=np.ones(50)
smoothing_filter/=len(smoothing_filter)
```

```python
onesecond_channel_0_smoothed = np.convolve(onesecond_channel_0, smoothing_filter, mode='same')
onesecond_channel_30_smoothed = np.convolve(onesecond_channel_30, smoothing_filter, mode='same')
onesecond_channel_59_smoothed = np.convolve(onesecond_channel_59, smoothing_filter, mode='same')
```

```python
fig, axs = plt.subplots(3)
axs[0].plot(onesecond_channel_0)
axs[0].plot(onesecond_channel_0_smoothed)

axs[0].set_title("1 sec of channel 0")
axs[1].plot(onesecond_channel_30)
axs[1].plot(onesecond_channel_30_smoothed)

axs[1].set_title("1 sec of channel 30")
axs[2].plot(onesecond_channel_59)
axs[2].plot(onesecond_channel_59_smoothed)

axs[2].set_title("1 sec of channel 59")

fig.legend(['data', 'smoothed data'], loc='upper right', bbox_to_anchor=(1.2, 0.5))

fig.tight_layout()

plt.show()
```

```python
nsamps = onesecond_channel_0.shape[0]

# want these initializations clarified
t = np.linspace(0,nsamps/sfreq,nsamps,endpoint=False)
f = np.linspace(0,sfreq,nsamps,endpoint=False)

onesecond_channel_0_sepctrum = np.fft.fft(onesecond_channel_0)
onesecond_channel_0_smoothed_sepctrum = np.fft.fft(onesecond_channel_0_smoothed)

onesecond_channel_30_sepctrum = np.fft.fft(onesecond_channel_30)
onesecond_channel_30_smoothed_sepctrum = np.fft.fft(onesecond_channel_30_smoothed)

onesecond_channel_59_sepctrum = np.fft.fft(onesecond_channel_59)
onesecond_channel_59_smoothed_sepctrum = np.fft.fft(onesecond_channel_59_smoothed)
```

```python
fig, axs = plt.subplots(3,2)

# want this formula clarified 
axs[0,0].plot(f,20*np.log10(1e-3+np.abs(onesecond_channel_0_sepctrum)))
axs[0,1].plot(f,20*np.log10(1e-3+np.abs(onesecond_channel_0_smoothed_sepctrum)))

axs[1,0].plot(f,20*np.log10(1e-3+np.abs(onesecond_channel_30_sepctrum)))
axs[1,1].plot(f,20*np.log10(1e-3+np.abs(onesecond_channel_30_smoothed_sepctrum)))

axs[2,0].plot(f,20*np.log10(1e-3+np.abs(onesecond_channel_59_sepctrum)))
axs[2,1].plot(f,20*np.log10(1e-3+np.abs(onesecond_channel_59_smoothed_sepctrum)))

plt.show()
fig.tight_layout()
```
