```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.figure
```

```python
fs = 8000
freq = 120
w = 2 * np.pi * freq / fs
nsamps = 1000
f = np.linspace(0,fs, nsamps, endpoint = False)
n = np.arange(nsamps)
x_n = np.cos(w*n)+0.5*np.cos(.3*w*n)
plt.plot(n,x_n)
```

```python
fft = np.fft.fft(x_n)
mag_fft = np.abs(fft)
```

```python
plt.plot(f, mag_fft)
plt.show()
argmax = np.argmax(mag_fft)
```
